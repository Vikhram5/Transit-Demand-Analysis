```python
import pandas as pd

dataframes = []

for i in range(1, 8):
    date = f"{i:02}" 
    df = pd.read_csv(f'Match/unique_unmatched_stops_{date}.csv')
    dataframes.append(df)

combined_df = pd.concat(dataframes)
combined_df = combined_df.drop_duplicates()
combined_df.to_csv('Match/combined_unmatched_stops_01-to-07.csv', index=False)

combined_df
```
